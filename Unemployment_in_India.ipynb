```python
import pandas as pd
import calendar

# Convert 'Date' column to datetime
df[' Date'] = pd.to_datetime(df[' Date'], dayfirst=True)

# Extract the month number
df['month_int'] = df[' Date'].dt.month

# Map the month number to the month abbreviation
df['month'] = df['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
top_Region = ir.sort_values(by=' Estimated Unemployment Rate (%)', ascending=False)
top_Region = top_Region[['Region', ' Estimated Unemployment Rate (%)']].drop_duplicates(subset='Region')
top_10_Region = top_Region.head(10)
top_10_Region.reset_index(drop=True, inplace=True)
top_10_Region[['Region', ' Estimated Unemployment Rate (%)']]
```
